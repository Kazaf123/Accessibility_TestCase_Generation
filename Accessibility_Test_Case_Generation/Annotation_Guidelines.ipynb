* [cite_start]**实时验证：** 必须在测试设备上开启屏幕阅读器（如TalkBack），对所构建的用例步骤进行实时验证，确保问题真实存在且交互符合辅助工具的操作逻辑 [cite: 90]。

## 3. 测试失败根因分析标注规范（交叉复核机制）

[cite_start]针对自动生成脚本在执行阶段失败的情况，为保证原因分析的客观性与可靠性，需采用严格的交叉复核（Cross-Validation）标注机制 [cite: 242]。

### 3.1 标注流程
1. [cite_start]**基准制定：** 标注前，两名独立研究人员需基于初始观察共同制定失败原因的分类基准 [cite: 242]。
2. [cite_start]**独立标注：** 两人独立检查所有未复现成功的用例及其执行日志，判断失败根因并进行归类 [cite: 242]。
3. [cite_start]**一致性检验：** 完成独立标注后，使用科恩卡帕系数 (Cohen's Kappa) 对两人的结果进行一致性检验（需达到高度一致性，如 Kappa > 0.82） [cite: 242]。

### 3.2 分歧解决机制
对于存在分歧的案例，必须启动第三方介入机制：
* [cite_start]由第三位研究人员介入 [cite: 243]。
* [cite_start]三人共同回放测试执行的完整日志 [cite: 243]。
* [cite_start]对比单帧GUI截图与控件树（XML Hierarchy）特征进行深入讨论 [cite: 243]。
* [cite_start]直至对根因达成完全一致的共识 [cite: 243]。

### 3.3 失败根因分类标签
标注人员需将失败原因严格归入以下三类之一：
* [cite_start]**界面结构复杂或动态布局问题：** 目标控件处于高度动态的界面环境（如嵌套滚动列表、轮播组件），导致GUI截图和控件树在单帧捕获中难以完整表达控件状态和交互路径，模型生成操作序列时错过关键控件 [cite: 243, 244]。
* [cite_start]**辅助功能语义缺失问题：** 控件缺少辅助功能语义标签（如 `contentDescription` 为空或角色信息不完整），导致模型在推理用户操作意图及辅助功能预期输出时面临语义缺失，未能构建正确的交互序列 [cite: 243, 245]。
* [cite_start]**动态内容加载或界面状态变化问题：** App存在动态内容加载、动画效果或页面跳转延迟。若测试脚本在界面完全渲染前执行，目标控件尚未出现在界面层级中，导致操作无效或无法触发预期问题 [cite: 243, 246]。
